# Operations On Fractions Using OOPS

```python
import math

class Fraction:

  def __init__(self,n,d):
    self.num = n
    self.den = d

  def __str__(self):
    self.simplify(self.num , self.den)
    return "{}/{}".format(self.num, self.den)


  def _simplify(self,p,q):
    m =  math.gcd(p,q)
    return p // m, q //m

  def __add__(self,other):
    temp_num = self.num*other.den + other.num*self.den
    temp_den = self.den*other.den

    temp_num , temp_den = self._simplify(temp_num, temp_den)
    return "{}/{}".format(temp_num, temp_den)

  def __sub__(self,other):
    temp_num = self.num*other.den - other.num*self.den
    temp_den = self.den*other.den
    temp_num , temp_den = self._simplify(temp_num, temp_den)
    return "{}/{}".format(temp_num, temp_den)


  def __mul__(self,other):
    temp_num = self.num*other.num
    temp_den = self.den*other.den
    temp_num , temp_den = self._simplify(temp_num, temp_den)
    return "{}/{}".format(temp_num , temp_den )

  def __truediv__(self,other):
    temp_num = self.num*other.den
    temp_den = self.den*other.num
    num , den = self._simplify(temp_num, temp_den)
    return "{}/{}".format(num , den)




while True:
    num1 = int(input("Enter the numerator of the  Fraction:"))
    den1 = int(input("Enter the Denominator of the  fraction:"))
    x = Fraction(num1,den1)
    print("Here's the Fraction : " , x)
    check = input("To do any Operation press 1 else 0:")
    if check == '1':
      num2 = int(input("Enter the numerator of the 2nd Fraction:"))
      den2 = int(input("Enter the Denominator of the 2nd fraction:"))
      y = Fraction(num2,den2)
      print("Here's the Fraction : " ,y)
      c = input("Choose the Operation + , - , * , / :")
      if c == "+":
        print( x ,"+", y ,"=" , x + y)
      elif c == "-":
        print( x ,"-", y ,"=" , x - y)
      elif c == "*":
        print( x ," *", y ,"=" , x * y)
      elif c == '/':
        print( x ," / ", y ,"=" , x / y)
      else:
        print("Invalid Choice")
      k = input("To exit press 0 else 1: ")
      if k == '0' :
        break;
    else:
      break;

```
